Question 1: What is hypothesis testing in statistics?

ANS - Hypothesis testing in statistics is a **method used to make decisions or draw conclusions about a population based on sample data**.

1. **State hypotheses**

   * **Null hypothesis (H₀):** A statement that there is *no effect* or *no difference*. (e.g., “The new drug has no effect compared to the old one.”)
   * **Alternative hypothesis (H₁ or Ha):** A statement that contradicts H₀, claiming there *is* an effect or difference.

2. **Choose a significance level (α):**

   * Common values: 0.05, 0.01.
   * This represents the probability of rejecting H₀ when it is actually true (Type I error).

3. **Collect data and compute a test statistic:**

   * Example: *t-test, z-test, chi-square test* depending on the situation.

4. **Find the p-value or compare with critical value:**

   * **p-value:** Probability of getting the observed results (or more extreme) if H₀ is true.
   * If **p ≤ α**, reject H₀. If **p > α**, fail to reject H₀.

5. **Make a decision:**

   * Reject H₀ → evidence supports H₁.
   * Fail to reject H₀ → not enough evidence against H₀.


Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?
ANS -

### **1. Null Hypothesis (H₀):**

* It is the **default assumption** or a statement of *no effect, no difference, or no relationship*.
* It suggests that any difference observed in the data is just due to **chance**.
* Example: *“The new teaching method has the same effect as the old method.”*

---

### **2. Alternative Hypothesis (H₁ or Ha):**

* It is the **opposite of the null hypothesis**.
* It states that there **is** an effect, difference, or relationship.
* Example: *“The new teaching method is more effective than the old method.”*

---

### **Key Differences**

| Aspect       | Null Hypothesis (H₀)              | Alternative Hypothesis (H₁)     |
| ------------ | --------------------------------- | ------------------------------- |
| Meaning      | No effect / No difference         | There is an effect / difference |
| Assumed True | Always assumed true at the start  | Tested against H₀               |
| Goal         | To try to reject it with evidence | To provide evidence in favor    |
| Example      | “The coin is fair”                | “The coin is biased”            |

---

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

ANS -

---

### **Significance Level (α) in Hypothesis Testing**

* The **significance level** (denoted by α) is the **threshold probability** used to decide whether to reject the null hypothesis (H₀).
* It represents the **risk of making a Type I error**, i.e., rejecting H₀ when it is actually true.

---

### **Common Values of α**

* **0.05 (5%)** → most common (5% chance of wrongly rejecting H₀).
* **0.01 (1%)** → stricter test, used in medical or critical research.
* **0.10 (10%)** → more lenient, used in exploratory studies.

---

### **Role in Deciding the Outcome**

1. **Set α before the test** (e.g., α = 0.05).
2. After computing the **p-value** from sample data:

   * If **p ≤ α** → Reject H₀ (evidence supports H₁).
   * If **p > α** → Fail to reject H₀ (not enough evidence against H₀).

---

### **Example**

Suppose a company claims their battery lasts **10 hours**. You test it with α = 0.05.

* If your test gives **p = 0.03** → Since 0.03 < 0.05, reject H₀ (evidence battery life ≠ 10 hours).
* If your test gives **p = 0.08** → Since 0.08 > 0.05, fail to reject H₀ (no strong evidence against the claim).

---

Question 4: What are Type I and Type II errors? Give examples of each.

ANS -
---

### **1. Type I Error (False Positive)**

* Occurs when we **reject the null hypothesis (H₀) when it is actually true**.
* It’s like a **false alarm**.
* Probability of making a Type I error = **α (significance level)**.

✅ **Example:**

* Court trial: H₀ = “The defendant is innocent.”
* Type I error = convicting an innocent person.
* In medicine: saying a drug works when it actually doesn’t.

---

### **2. Type II Error (False Negative)**

* Occurs when we **fail to reject the null hypothesis (H₀) when it is actually false**.
* It’s like **missing a real effect**.
* Probability of making a Type II error = **β**.
* Power of a test = (1 – β).

✅ **Example:**

* Court trial: H₀ = “The defendant is innocent.”
* Type II error = letting a guilty person go free.
* In medicine: saying a drug doesn’t work when it actually does.

---

### **Comparison Table**

| Error Type      | What happens?                      | Real-world analogy            |
| --------------- | ---------------------------------- | ----------------------------- |
| **Type I (α)**  | Reject H₀ when H₀ is true          | Convicting an innocent person |
| **Type II (β)** | Fail to reject H₀ when H₀ is false | Freeing a guilty person       |

---

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

ANS -
Z-test vs. T-test
Aspect	Z-test	T-test
Population variance (σ²)	Known	Unknown
Sample size (n)	Large (n ≥ 30, Central Limit Theorem applies)	Small (n < 30)
Distribution used	Standard Normal Distribution (Z)	Student’s t Distribution
Shape of distribution	Fixed (normal curve)	Depends on sample size (wider tails for small n)
Purpose	Compare sample mean to population mean, or compare proportions	Compare sample mean(s) when population variance is unknown
Examples of use	Checking if the average height of students differs from a known population mean when σ is known	Testing if a new teaching method improves scores when σ is unknown and sample is small
When to Use

Use a Z-test when:

Population variance (σ²) or standard deviation (σ) is known.

Sample size is large (n ≥ 30).

Use a T-test when:

Population variance is unknown.

Sample size is small (n < 30).

✅ Example:

A factory claims the mean weight of chocolate bars is 100g.

If you know the population standard deviation (σ = 2g) and sample 50 bars → Z-test.

If σ is unknown and you only have 15 bars → T-test.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Hypothesized population mean
mu_0 = 50

# Assume known population standard deviation (σ)
sigma = 1

# Sample statistics
sample_mean = np.mean(data)
n = len(data)

# Compute Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Compute two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Output
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.(Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # population mean and std dev
n = 100             # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
alpha = 0.05

# Step 3: Calculate 95% confidence interval for mean using T-distribution
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"95% Confidence Interval = ({ci_lower:.4f}, {ci_upper:.4f})")

# Step 5: Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.(Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

# Sample dataset (normally distributed)
np.random.seed(42)
data = np.random.normal(50, 5, 100)  # mean=50, std=5, n=100

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first 10 Z-scores for demonstration
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, color='lightcoral', edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.axvline(0, color='blue', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.axvline(1, color='green', linestyle='dashed', linewidth=2, label="1 Std Dev")
plt.axvline(-1, color='green', linestyle='dashed', linewidth=2)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
